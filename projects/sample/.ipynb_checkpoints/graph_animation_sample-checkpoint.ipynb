```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
ITV = 100 # インターバル。値を下げるほど FPS が上がる
```

```python
fig, ax = plt.subplots()  # 図と軸を作成
xdata, ydata = [], []  # x軸とy軸のデータを空にする
ln, = plt.plot([], [], 'ro')  # 赤い丸をプロットする線を作成

def init():  # 初期化関数
    ax.set_xlim(0, 2 * np.pi)  # x軸の範囲を設定
    ax.set_ylim(-1, 1)  # y軸の範囲を設定
    return ln,  # プロットした線を返す

def update(frame):  # 更新関数
    xdata.append(frame)  # x軸データにframeを追加
    ydata.append(np.sin(frame))  # y軸データにsin(frame)を追加
    ln.set_data(xdata, ydata)  # プロットした線のデータを更新
    return ln,  # 更新した線を返す

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2 * np.pi, 128),
                    init_func=init, blit=True, interval=ITV)  # アニメーションを作成

HTML(ani.to_html5_video())  # アニメーションをHTML5ビデオとして表示
```

```python
fig, ax = plt.subplots()  # 図と軸を作成
x = np.linspace(0, 2 * np.pi, 100)  # 0から2πまでの間を100分割した値をxに格納
y = np.random.rand(100) * 2 - 1  # -1から1の間のランダムな値を100個生成してyに格納
ln, = plt.plot(x, y)  # xとyの値でプロットした線を作成

def update(frame):  # 更新関数
    y = np.random.rand(100) * 2 - 1  # -1から1の間のランダムな値を100個生成してyに格納
    ln.set_ydata(y)  # プロットした線のy軸のデータを更新
    return ln,  # 更新した線を返す

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2 * np.pi, 128),
                    blit=True, interval=ITV)  # アニメーションを作成

HTML(ani.to_html5_video())  # アニメーションをHTML5ビデオとして表示
```

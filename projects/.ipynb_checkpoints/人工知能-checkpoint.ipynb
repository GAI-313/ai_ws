## Q1. 1~10 までの和を for 文で求めるプログラム

````python
s = 0
for i in range(1,11):
    s = s + i
print(f"Result sum = {s}")
````

### 解説
　まずはじめに初期値となる変数 `s` を定義する。今後この変数に $1$ ~ $10$ までの値が順番に足されていくプログラムを実装するためである。<br>
　2行目の
```python
for i in range(1,11):
```
　が `for` 文の本体である。これは
```
range(1,11) から得た値を i に代入して処理を繰り返していく
```
という意味である。この `for` 文に含まれる関数 `range` は以下の意味を持つ。
```python
range(start, stop, step)
```
　`start` の部分にカウントの始めの数字を定義し、 `stop` にカウントを終了する数字を定義する。そして `step` にカウントの間隔を設定することで **`start` から `stop` までの範囲を `step` 刻みでカウントする** 処理を実装できる。<br>
　ちなみにこの `step` は省略可能で、上記のコードのように `step` を定義しないとデフォルトで1づつカウントしてくれる。<br>
　ここで疑問に思っただろうが、終了値 `stop` に $11$ を設定しているのに、変数 `i` には $10$ までしか代入されない。これは `range` 関数の使用で、**`stop` 値未満の値までを範囲として扱う** というルールに基づいているからである。<br>

---
　次に、`for` 文内の処理の内容を解説する。ここでは
```python
s = s + i
```
という処理が実装されている。ここでは変数 `s` に変数 `i` を加算する処理を行なっており、これが `for` 文内で行われていることからカウントされた値が常に `s` に代入される処理が実行される。

---

　最後に `print` 関数を使って結果を出力しているのだが、ここでは **F文字列** という特殊な文字列を用いて変数を展開している。
```python
print(f"Result s = {s}")
```
F文字列を活用することで明示的に変数を展開することができ、型変換や書式演算子など少々高度な作業の手間を省くことができる。しかしこの方法は他の展開方法と比べてリソースを多く消費するため、多用はお勧めしない。

---

　以下のコードは別解として、**内包表記** を用いて1行で課題を遂行する方法である。この手法を用いる場合変数名に `sum` を用いると合計値を求める関数 `sum` が動作しないので変数名に気をつけよう。


````python
print(f"result sum = {sum([i for i in range(1,11)])}")
````

## Q2. 1~10 までの和を while 文で求めるプログラム

````python
i , s = 0, 0
while i <= 10:
    s = s + i
    i += 1
print(f"result sum {s}")
````

### 解説
　コードのはじめで 1行で 2変数の初期値を設定している。カンマで区切って定義したい変数と値分書くことで簡単に1行に複数変数をまとめることが可能である。
```python
i , s = 0, 0
```
　次の `while` 文では以下の条件式を持つ。
```python
i <= 10
```
　これは変数 `i` が $10$ 未満の場合は `True` であることを示しており、これを`while` 文に記述することで変数 `i` が $10$ 未満の間は `while` 文内の処理を実行する処理が実装できる。<br>
　`while` 文内では変数 `s` に `i` を加算し、`i` に $1$ を加算する処理を実装している。これにより `while` 文内で常に `s` と `i` が加算され続ける処理が実装される。<br>
　最後は合計値 `s` を表示して終了である。

## Q3. ユーザー定義関数

````python
def for_loop(count): return sum([i for i in range(1,count + 1)])

def while_loop(count):
    i , s = 0, 0
    while i <= count:
        s = s + i
        i += 1
    return s

value = 10

print(f"for_loop func result : {for_loop(value)}")
print(f"while_loop func result : {while_loop(value)}")
````

### 解説
　単純に前回書いたコードを関数にまとめただけである。関数の定義方法は以下の通りである。
```python
def 関数名 (引数1, 引数2, ...):
    # 関数内の処理
```
　ここでは `for_loop` と `while_loop` の2つの関数を定義しているが、どちらも
$$
f(x) = SUM(x)
$$
である。

## Q1. ⻑さ10のリストを作成し、1〜10を代入せよ。 またWhile文を用いて、リストの要素の合計を求めよ

````python
list = [ i for i in range(1,11)] # 1 ~ 10 のリストを作成
i, s = 0, 0
while i <= len(list):
    s = s + i
    i += 1
print(f"result sum = {s}")
````

## Q2. ⻑さ10のリストを作成し、1〜10を代入せよ。 またfor文を用いて、リストの要素の合計を 求めよ

````python
list = [ i for i in range(1,11)] # 1 ~ 10 のリストを作成
print(f"result sum = {sum(i for i in list)}") # for 文で求める
````

## Q3. ⻑さ10のリストを作成し、1〜10を代入せよ。ユーザー定義関数にそのリストを渡し、 ユーザー定義関数にて和と積を求めよ。また結 果はリスト[和,積]として戻し、ユーザー定義 関数の外で表示せよ

````python
def sum_prid(nums):
    s = 1
    for i in nums:
        s = s * i
    return sum(nums), s

list = [ i for i in range(1,11)] # 1 ~ 10 のリストを作成
s, p = sum_prid(list)

print(f"SUM = {s}")
print(f"PROD = {p}")
````

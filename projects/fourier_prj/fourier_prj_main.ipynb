```python
import numpy as np
import matplotlib.pyplot as plt

# 新しい関数 f(t) の定義
def f(t):
    return np.piecewise(t, [t <= 0, t > 0], [lambda x: x + np.pi, lambda x: np.pi - x])

T = 2 * np.pi
N = 40
dt = 0.001

t = np.arange(-T / 2, T / 2, dt)

# フーリエ係数 c_k の新しい計算式
c_k_list = [(1-(-1)**k) / (np.pi * k**2) for k in range(1, N)]

# c_0 の値を更新
c_0 = np.pi / 2

# 近似関数の計算
f_approx = np.ones_like(t) * c_0
for (c_k, k) in zip(c_k_list, range(1, N)):
    f_approx += 2 * (c_k * np.exp(2j * np.pi * t * k / T)).real

# 関数 f(t) のグラフ
plt.plot(t, f(t))

# 近似関数のグラフ
plt.plot(t, f_approx)

plt.legend()
plt.show()
```
